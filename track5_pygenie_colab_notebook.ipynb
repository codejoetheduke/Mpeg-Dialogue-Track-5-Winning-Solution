```python
%%writefile decoder.py
# decoder.py
import os
import tempfile
from pygenie import run

def decode_mgb_to_fastq(mgb_path):
    """Decode a single .mgb file to a temporary FASTQ file."""
    import tempfile, os, subprocess
    tmpdir = tempfile.mkdtemp()
    fastq_path = os.path.join(tmpdir, os.path.splitext(os.path.basename(mgb_path))[0] + ".fastq")
    try:
        run(input_file=mgb_path, output_file=fastq_path, force=True)
        return fastq_path
    except subprocess.CalledProcessError as e:
        print(f"‚ö†Ô∏è Genie failed on {mgb_path}: {e}")
        return None
```

```python
%%writefile processor.py
# processor.py
from Bio import SeqIO
from collections import Counter
import os

def process_file_combined(file_path, N=100):
    """Stream process FASTQ to avoid memory overload."""
    try:
        c = Counter()
        total_len = 0
        total_reads = 0

        A_count = T_count = G_count = C_count = 0

        for record in SeqIO.parse(str(file_path), "fastq"):
            seq = str(record.seq).upper()
            total_reads += 1
            total_len += len(seq)
            c[seq] += 1

            A_count += seq.count("A")
            T_count += seq.count("T")
            G_count += seq.count("G")
            C_count += seq.count("C")

        if total_reads == 0:
            return None

        features = {
            "filename": os.path.basename(file_path),
            "num_reads": int(total_reads),
            "avg_read_len": total_len / total_reads,
            "gc_content": (G_count + C_count) / total_len,
            "A": A_count / total_len,
            "T": T_count / total_len,
            "G": G_count / total_len,
            "C": C_count / total_len,
        }

        # Add top-N reads
        top_reads = [seq for seq, _ in c.most_common(N)]
        while len(top_reads) < N:
            top_reads.append("NA")

        for i in range(N):
            features[f"rank_{i+1}_read"] = top_reads[i]

        features["total_reads"] = total_reads
        features["duplicates"] = total_reads - len(c)

        return features

    except Exception as e:
        print(f"‚ùå Error processing {file_path}: {e}")
        return None
```

```python
%%writefile parallel_utils.py
import os
from pathlib import Path
import pandas as pd
from concurrent.futures import ProcessPoolExecutor, as_completed
from config import num_workers
from processor import process_file_combined
import pyfiglet
from rich import print

def split_reads_counts(df, N, prefix="Train", append=True):
    """Split df into reads and counts CSVs with consistent filename column."""
    # Ensure a "filename" column exists
    if "filename" not in df.columns and "fastq_filename" in df.columns:
        df = df.rename(columns={"fastq_filename": "filename"}).copy()

    # Reads columns: filename + all rank_n_read
    reads_cols = ["filename"] + [f"rank_{i}_read" for i in range(1, N+1) if f"rank_{i}_read" in df.columns]
    df_reads = df[reads_cols].copy()

    # Counts columns: filename + everything else (excluding reads)
    counts_cols = ["filename"] + [c for c in df.columns if c not in reads_cols]
    df_counts = df[counts_cols].copy()

    # Drop problem columns if they exist
    for c in ["duplicates", "total_reads"]:
        if c in df_counts.columns:
            df_counts = df_counts.drop(columns=[c])

    reads_file = f"{prefix.lower()}_reads.csv"
    counts_file = f"MicroBio_{prefix}_Counts.csv"

    # Append or overwrite
    mode = 'a' if append else 'w'
    write_header = not (append and os.path.exists(reads_file))

    df_reads.to_csv(reads_file, index=False, mode=mode, header=write_header)
    df_counts.to_csv(counts_file, index=False, mode=mode, header=write_header)


def parallel_extract(path_list, N=100):
    files = []
    for d in path_list:
        if os.path.exists(d):
            files.extend(list(Path(d).glob("*.fastq")))
    results = []
    with ProcessPoolExecutor(max_workers=num_workers) as executor:
        futures = {executor.submit(process_file_combined, f, N): f for f in files}
        for i, future in enumerate(as_completed(futures), 1):
            res = future.result()
            if res:
                results.append(res)

    return pd.DataFrame(results)

def banner():
    banner = pyfiglet.figlet_format("EVER LEARNERS", font="slant")
    print(f"[bold cyan]{banner}[/]")
```

```python
import os
import shutil
import pandas as pd
from pathlib import Path
from sklearn.preprocessing import LabelEncoder
from rich import print
from tqdm import tqdm
from config import *
from decoder import decode_mgb_to_fastq
from parallel_utils import parallel_extract, split_reads_counts, banner


def process_folder(folder, prefix="Train", train_labels=None):
    import os
    """Decode and process MGB files in a folder."""
    mgb_files = list(Path(folder).glob("*.mgb"))

    for mgb_file in tqdm(mgb_files, desc=f"{prefix} samples", unit="sample"):
        # Decode into a temporary FASTQ
        fastq_path = decode_mgb_to_fastq(str(mgb_file))

        # Skip if decoding failed
        if fastq_path is None:
            print(f"‚ö†Ô∏è Skipping {mgb_file.name} (decode failed)")
            continue

        # Extract features
        df = parallel_extract([Path(fastq_path).parent], N)

        # --- Skip if no reads extracted ---
        if df is None or df.empty:
            print(f"‚ö†Ô∏è Skipping {mgb_file.name} (no features extracted)")
            try:
                os.remove(fastq_path)
                shutil.rmtree(os.path.dirname(fastq_path))
            except Exception as e:
                print(f"‚ö†Ô∏è Cleanup failed for {fastq_path}: {e}")
            continue

        # --- üîó Merge labels (TRAIN only) ---
        if prefix == "Train" and train_labels is not None:
            # Convert labels .mgb ‚Üí .fastq
            train_labels["fastq_filename"] = train_labels["filename"].str.replace(".mgb", ".fastq")

            # Normalize df's filename column
            if "filename" in df.columns:
                df["fastq_filename"] = df["filename"].apply(lambda x: Path(x).name)
            elif "path" in df.columns:
                df["fastq_filename"] = df["path"].apply(lambda x: Path(x).name)
            else:
                print(f"‚ö†Ô∏è No filename column in extracted df for {mgb_file.name}, skipping merge")
                df["fastq_filename"] = Path(fastq_path).name

            # Merge with labels
            df = df.merge(
                train_labels[["fastq_filename", "SampleType"]],
                on="fastq_filename",
                how="left"
            )

            # Encode categorical labels
            if "SampleType" in df.columns:
                df["label"] = LabelEncoder().fit_transform(df["SampleType"].astype(str))

        # Split into reads/counts CSVs
        split_reads_counts(df, N, prefix=prefix, append=True)

        # Cleanup temporary files
        try:
            os.remove(fastq_path)
            shutil.rmtree(os.path.dirname(fastq_path))
        except Exception as e:
            print(f"‚ö†Ô∏è Cleanup failed for {fastq_path}: {e}")



def delete_existing_csvs():
    """Remove any old MicroBio CSVs to start fresh."""
    for pattern in [
        "train_reads.csv", "MicroBio_Train_Counts.csv",
        "test_reads.csv", "MicroBio_Test_Counts.csv"
    ]:
        path = Path(pattern)
        if path.exists():
            path.unlink()


def main():
    import os
    banner()
    delete_existing_csvs()

    # Load train labels once
    train_labels = None
    if os.path.exists(TRAIN_LABELS_FILE):
        train_labels = pd.read_csv(TRAIN_LABELS_FILE)

    # # Process TRAIN
    # print("[bold cyan]üöÄ Processing TRAIN")
    # for folder in train_dirs:
    #     process_folder(folder, prefix="Train", train_labels=train_labels)

    # # Process TEST
    # print("[bold cyan]üöÄ Processing TEST")
    # for folder in test_dirs:
    #     process_folder(folder, prefix="Test")

    # Process VIOME
    print("[bold green]üöÄ Processing Viome")
    for folder in viome_dirs:
        process_folder(folder, prefix="Viome")

    print("[bold green]‚úÖ Pipeline completed!")


if __name__ == "__main__":
    main()
```
